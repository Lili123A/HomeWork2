```python
import array

def main():
    N = int(input("N="))
    a = array.array('i', [0] * 10)
    for i in range(1, N + 1):
        a[i - 1] = i * 2 - 1
    for i in range(1, N + 1):
        print(f"{i}: {a[i - 1]}")

if __name__ == "__main__":
    main()
```

```python
import array

def main():
    N = int(input("N= "))
    a1 = int(input("a1="))
    D = int(input("D="))
    arr = array.array('i')
    current_value = a1
    for i in range(N):
        arr.append(current_value)
        current_value *= D
    for i in range(N):
        print(f"{i + 1}: {arr[i]}")
if __name__ == "__main__":
    main()
```

```python
import array
def main():
    N = int(input("Введите N: "))
    a = array.array('i', [0] * 10)
    a[0] = 1
    a[1] = 1
    for i in range(2, N):
        a[i] = a[i - 2] + a[i - 1]
    for i in range(N):
        print(f"{i + 1}: {a[i]}")
if __name__ == "__main__":
    main()
```

```python
def main():
    M = int(input("M="))
    N = int(input("N="))
    print(f"Введите {M} чисел:")
    numbers = [int(input()) for _ in range(M)]
    matrix = []
    for _ in range(N):
        matrix.append(numbers[:])
    for row in matrix:
        print(' '.join(map(str, row)))
if __name__ == "__main__":
    main()
```

```python
def main():
    M = int(input("M="))
    N = int(input("N="))
    print(f"Введите {N} чисел для первой строки:")
    row = [int(input()) for _ in range(N)]
    matrix = [row[:]]
    for i in range(1, M):
        matrix.append(row[:])
    for row in matrix:
        print(' : '.join(map(str, row)), end=' : \n')
if __name__ == "__main__":
    main()
```

```python
def main():
    M = int(input("M="))
    N = int(input("N="))
    D = int(input("D="))
    print(f"Введите {M} чисел для первого столбца:")
    matrix = []
    for i in range(M):
        value = int(input(f"{i + 1}: "))
        row = [value]
        for j in range(1, N):
            row.append(row[j - 1] + D)
        matrix.append(row)
    for row in matrix:
        print(' : '.join(map(str, row)), ' : ')
if __name__ == "__main__":
    main()
```
