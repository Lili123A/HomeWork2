```python
x1 = int(input("x1="))
x2 = int(input("x2="))
y1 = int(input("y1="))
y2 = int(input("y2="))
def RectPS(x1, y1, x2, y2, P, S):
    ширина = abs(x2 - x1)
    длина = abs(y2 - y1)
    P[0] = 2 * (ширина + длина)
    S[0] = ширина * длина
if __name__ == "__main__":
    x1, y1 = 0, 0
    x2, y2 = 4, 3
    P1 = [0]
    S1 = [0]
    RectPS(x1, y1, x2, y2, P1, S1)
    print(P1,S1)
    x1, y1 = -2, -1
    x2, y2 = 5, 6
    P2 = [0]
    S2 = [0]
    RectPS(x1, y1, x2, y2, P2, S2)
    print(P2,S2)
    x1, y1 = 3, 7
    x2, y2 = 3, 7
    P3 = [0]
    S3 = [0]
    RectPS(x1, y1, x2, y2, P3, S3)
    print(P3,S3)
```

```python
A = int(input("A="))
B = int(input("B="))
С = int(input("С="))
def RootCount(A, B, C):
    D = B**2 - 4 * A * C
    if D > 0:
        return 2
    elif D == 0:
        return 1
    else:
        return 0
if __name__ == "__main__":
    equations = [(1, -3, 2),(1, 2, 1),(1, 1, 2)]
    for i, (a, b, c) in enumerate(equations, start=1):
        roots = RootCount(a, b, c)
        print(roots)
```
