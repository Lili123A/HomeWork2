```python
n = int(input("n="))
def f(n):
    r = []
    for x in range(1, n + 1):
        d = [int(c) for c in str(x)]
        p = True
        for q in d:
            if q == 0 or x % q != 0:
                p = False
                break
        if p:
            r.append(x)
    return r
print( f(n))
```

```python
import math
a, b = map(int, input("(A/B)=").split('/'))
c, d = map(int, input("(C/D)=").split('/'))
def f(a, b, c, d):
    n = a * d - c * b
    m = b * d
    g = math.gcd(n, m)
    return (n // g, m // g)
result = f(a, b, c, d)
print(f"Результат вычитания= {result[0]}/{result[1]}")
```

```python
S1 = int(input("S1="))
S2 = int(input("S2="))
S1 = 1
S2 = 1
day = 2
while True:
    day += 1
    Sn = (S1 + S2) / 2 + 2
    if Sn >= 50:
        break
    S1 = S2
    S2 = Sn
print(f"Силы Незнайки иссякнут через {day} дней.")
```
